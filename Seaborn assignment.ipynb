### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
#Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five common plots that can be created using Seaborn along with their uses:
```

```python
#Scatter Plot:
#Use: Scatter plots are used to visualize the relationship between two continuous variables. They are helpful in identifying patterns, correlations, or clusters in the data.

#Line Plot:
#Use: Line plots are useful for visualizing the trend or pattern of one or more continuous variables over time or any other ordered dimension. They are often used for time series data.
#Bar Plot:

#Use: Bar plots are used to represent categorical data with rectangular bars. They are suitable for comparing the values of different categories and identifying the highest or lowest values easily.

#Histogram:
#Use: Histograms are used to display the distribution of a single numeric variable. They show the frequency of data points falling into various bins or intervals, helping to understand the data's underlying distribution.

#Box Plot (Box-and-Whisker Plot):
#Use: Box plots are used to display the distribution and identify outliers in the data. They show the median, quartiles, and potential outliers in a visually concise manner, making it easy to compare different groups or categories.
```

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri = sns.load_dataset('fmri')
```

```python
sns.lineplot(data = fmri , x = "timepoint", y = "signal" , hue = "event" , style = "region")
plt.xlabel("Timepoint")
plt.ylabel("signal")
plt.title("Line Plot of Signal over Time for Different Events and Regions")
plt.show()
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
titanic_data = sns.load_dataset('titanic')
```

```python
plt.figure(figsize=(10,6))
#box plot for age vs pclass
plt.subplot(1,2,1)
sns.boxplot(data=titanic_data, x="pclass", y="age")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.title("Box Plot of Age vs. Passenger Class")

# Box plot for 'fare' vs. 'pclass'
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x="pclass", y="fare")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.title("Box Plot of Fare vs. Passenger Class")

plt.tight_layout()

plt.show()
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds_data = sns.load_dataset('diamonds')
sns.histplot(data = diamonds_data, x = "price" , hue = "cut")
```

### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
iris_data = sns.load_dataset("iris")
sns.pairplot(data = iris_data , hue = "species")
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights_long = sns.load_dataset("flights")
flights = flights_long.pivot("month", "year", "passengers")

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax)
```
